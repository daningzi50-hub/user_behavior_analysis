```python
import random
input_file='UserBehavior.csv'
output_file='sample.csv'
sample_size=100_0000

with open(input_file,'r',encoding='utf-8') as f:
    total=sum(1 for _ in f)
prob=sample_size/total


with open(input_file,'r',encoding='utf-8') as fin,open(output_file,'w',encoding='utf-8') as fout:
   
    for line in fin:
        if random.random()<prob:
            fout.write(line)
    
```

```python
import random

input_file = 'UserBehavior.csv'
output_file = 'sample.csv'
sample_users = 100000  # 抽10万用户

# 第一遍：收集所有user_id
user_ids = set()
with open(input_file, 'r', encoding='utf-8') as f:
   
    for line in f:
        uid = line.split(',')[0]  # user_id在第一列
        user_ids.add(uid)

# 随机抽取10万个user_id
selected = set(random.sample(list(user_ids), sample_users))

# 第二遍：写入这些用户的所有行
with open(input_file, 'r', encoding='utf-8') as fin, \
     open(output_file, 'w', encoding='utf-8') as fout:
    
    
    for line in fin:
        uid = line.split(',')[0]
        if uid in selected:
            fout.write(line)
```
